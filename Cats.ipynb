Какие номера у двух предложений, ближайших к нему по этому расстоянию (строки нумеруются с нуля)? Эти два числа и будут ответами на задание. Разумеется, использованный вами метод крайне простой. Например, он не учитывает формы слов (так, cat и cats он считает разными словами, хотя по сути они означают одно и то же), не удаляет из текстов артикли и прочие ненужные слова. Позже мы будем подробно изучать анализ текстов, где выясним, как достичь высокого качества в задаче поиска похожих предложений.

```python
import re
import numpy as np

sent_list = []

#создаем список предложений, каждое предложение - список слов
with open('cats_sentences.txt', 'r') as file_obj:
    for line in file_obj:
        sent_list.append(re.split('[^a-z]', line.lower()))

#удаляем пустые слова
for s in sent_list: 
    while '' in s:   
        s.remove('')
    

#создаем словарь токенов с уникальными индексами
tokens = {}

index = 0
for line in sent_list:
     for word in line:
        if tokens.has_key(word)==False:
            tokens[word] = index
            index = index + 1
 
 
#создаем матрицу заполняем ее
cats_matrix = np.zeros((len(sent_list),len(tokens))) 

for index,line in enumerate(sent_list):
    for k,v in tokens.items():
        a = line.count(k)
        cats_matrix[index][v] = a
     
        
#вычисляем косинусное расстояние
from scipy.spatial import distance
cl = {}
for i in range(1,len(cats_matrix)):
   cl[i] =distance.cosine(cats_matrix[0],cats_matrix[i])

sorted(cl.items(),key=lambda x:x[1])
```
